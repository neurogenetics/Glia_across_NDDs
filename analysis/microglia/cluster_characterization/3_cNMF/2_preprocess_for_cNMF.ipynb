```python
import os
import pandas as pd
import numpy as np
from scipy.io import mmread
import scipy.sparse as sp
import matplotlib.pyplot as plt
from IPython.display import Image
import scanpy as sc
from cnmf import cNMF, Preprocess
import anndata
import seaborn as sns
```

```python
adata_AMPPD = sc.read_h5ad("/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/microglia_cNMF_AMPPD.h5ad")
adata_Gerrits = sc.read_h5ad("/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/microglia_cNMF_Gerrits.h5ad")
adata_Mathys = sc.read_h5ad("/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/microglia_cNMF_Mathys.h5ad")
adata_Pineda = sc.read_h5ad("/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/microglia_cNMF_Pineda.h5ad")
```

```python
# transfer raw counts to new objects to be processsed by scanpy
adata_new_AMPPD = anndata.AnnData(X=adata_AMPPD.X, obs=adata_AMPPD.obs.copy(), var=adata_AMPPD.var.copy())
adata_new_Gerrits = anndata.AnnData(X=adata_Gerrits.X, obs=adata_Gerrits.obs.copy(), var=adata_Gerrits.var.copy())
adata_new_Mathys = anndata.AnnData(X=adata_Mathys.X, obs=adata_Mathys.obs.copy(), var=adata_Mathys.var.copy())
adata_new_Pineda = anndata.AnnData(X=adata_Pineda.X, obs=adata_Pineda.obs.copy(), var=adata_Pineda.var.copy())
```

```python
# batch correction
p = Preprocess(random_seed=14)
(adata_c_AMPPD, adata_tp10k_AMPPD, hvgs_AMPPD) = p.preprocess_for_cnmf(adata_new_AMPPD, harmony_vars=['donor'], n_top_rna_genes=2000, librarysize_targetsum=1e5,
                                                                       max_scaled_thresh=None, quantile_thresh=.9999, makeplots=True,
                                                                       save_output_base='/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/AMPPD/microglia')
(adata_c_Gerrits, adata_tp10k_Gerrits, hvgs_Gerrits) = p.preprocess_for_cnmf(adata_new_Gerrits, harmony_vars=['donor'], n_top_rna_genes=2000, librarysize_targetsum=1e5,
                                                                             max_scaled_thresh=None, quantile_thresh=.9999, makeplots=True,
                                                                             save_output_base='/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/Gerrits/microglia')
(adata_c_Mathys, adata_tp10k_Mathys, hvgs_Mathys) = p.preprocess_for_cnmf(adata_new_Mathys, harmony_vars=['donor'], n_top_rna_genes=2000, librarysize_targetsum=1e5,
                                                                          max_scaled_thresh=None, quantile_thresh=.9999, makeplots=True,
                                                                          save_output_base='/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/Mathys/microglia')
(adata_c_Pineda, adata_tp10k_Pineda, hvgs_Pineda) = p.preprocess_for_cnmf(adata_new_Pineda, harmony_vars=['donor'], n_top_rna_genes=2000, librarysize_targetsum=1e5,
                                                                          max_scaled_thresh=None, quantile_thresh=.9999, makeplots=True,
                                                                          save_output_base='/data/ADRD/glia_across_NDDs/analysis/microglia/cluster_characterization/cNMF_by_dataset/Pineda/microglia')
```
